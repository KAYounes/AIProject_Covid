# Normalizing 'location' and 'Country' Columns

## Z-Score Normalization

```python
z_score_location_col = original_df['location']
z_score_location_col = (z_score_location_col - z_score_location_col.mean()) / (z_score_location_col.std())
print('min ', z_score_location_col.min(), '\nmax ', z_score_location_col.max(), '\nmean ', z_score_location_col.mean(), '\nstd ', z_score_location_col.std())
```

```python
z_score_country_col = original_df['country']
z_score_country_col = (z_score_country_col - z_score_country_col.mean()) / (z_score_country_col.std())
print('min ', z_score_country_col.min(), '\nmax ', z_score_country_col.max(), '\nmean ', z_score_country_col.mean(), '\nstd ', z_score_country_col.std())
```

```python
z_score_location_country = original_df.copy()
z_score_location_country['location'] = z_score_location_col
z_score_location_country['country'] = z_score_country_col
z_score_location_country.to_csv(r'\Users\KAYEG\Documents\ASU\FALL - 21 - Term 7\Machine Learning\Project\Project\datas\z_score_location_country.csv', index=False)
```

## min-max Normalization

```python
min_max_location_col = original_df['location']
min_max_location_col = (min_max_location_col - min_max_location_col.min()) / (min_max_location_col.max() - min_max_location_col.min())
print('min ', min_max_location_col.min(), '\nmax ', min_max_location_col.max(), '\nmean ', min_max_location_col.mean(), '\nstd ', min_max_location_col.std())
```

```python
min_max_country_col = original_df['country']
min_max_country_col = (min_max_country_col - min_max_country_col.min()) / (min_max_country_col.max() - min_max_country_col.min())
print('min ', min_max_country_col.min(), '\nmax ', min_max_country_col.max(), '\nmean ', min_max_country_col.mean(), '\nstd ', min_max_country_col.std())
```

```python
min_max_location_country = original_df.copy()
min_max_location_country['location'] = min_max_location_col
min_max_location_country['country'] = min_max_country_col
min_max_location_country.to_csv(r'\Users\KAYEG\Documents\ASU\FALL - 21 - Term 7\Machine Learning\Project\Project\datas\min_max_location_country.csv', index=False)
```

# One-Hot Encoding Location and Country Columns


```python
one_hot_location_country = pd.get_dummies(original_df, columns=['location', 'country'])
result_col = original_df['result']
one_hot_location_country.drop('result', axis=1, inplace=True)
one_hot_location_country['result'] = result_col
one_hot_location_country.to_csv(r'\Users\KAYEG\Documents\ASU\FALL - 21 - Term 7\Machine Learning\Project\Project\datas\one_hot_location_country.csv', index=False)
```
